# K-Means Clustering

### Full code

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

x, y, centers = make_blobs(n_samples = 500, centers = 4, cluster_std = 0.5, random_state = 0, return_centers = True)

kmeans = KMeans(n_clusters = 4)
kmeans.fit(x)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

plt.scatter(x[:,0], x[:,1], c = labels, s = 50, cmap = 'viridis')
plt.scatter(centroids[:,0], centroids[:,1], c = 'red', s = 200 , alpha = 1, marker = 'X')

plt.title("K-Means Clustering")
plt.show()
```

### Variable Explanation 

- ***KMeans -*** Build in functions that knows how to group the dataset into clusters
- ***make_blobs -*** Build in dataset whivh contains **n** number of data points
- ***x -*** Array of position of **n** number of datapoints
- ***y -*** Array of c
- ***n_samples -*** Total count of data points
- ***centers -*** Mention how many clusters are being formed
- ***cluster_std -*** Unit distance from the cluster center to the data point
- ***random_state -*** Reproductability value 
- ***cluster_centers_ -*** Array of centroids of all the clusters
- ***labels_ -*** Array of color scheme for all the datapoints
- ***c -*** Update color group number for each data point
- ***s -*** Size of each data point in the graph
- ***cmap -*** Map color with the group/cluster number
- ***alpha -*** Opacity of the marker

### _Evaluate the optimal number of clusters using the elbow method._

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

rating = np.random.uniform(1, 10, 1000)
votes = np.random.randint(1000, 3000000, 1000)
movies = []

for i in range(len(rating)):
    movies.append([rating[i], votes[i]])
    
x = np.array(movies)

kmeans = KMeans(n_clusters = 4)
kmeans.fit(x)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

plt.scatter(x[:,0], x[:,1], c = labels, s = 50, cmap = 'viridis')
plt.scatter(centroids[:,0], centroids[:,1], c = 'red', s = 200 , alpha = 1, marker = 'X')

plt.title("K-Means Clustering")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

rating = np.random.uniform(1, 10, 1000)
votes = np.random.randint(1000, 3000000, 1000)
movies = []

for i in range(len(rating)):
    movies.append([rating[i], votes[i]])
    
x = np.array(movies)

wcss = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

plt.plot(k_values, wcss, 'bx-')
plt.title('Elbow Method to Determine Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS (Inertia)')
plt.show()
```
