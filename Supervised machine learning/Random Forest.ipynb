# Random Forest

### Full Code

```python
import pandas as pd
from sklearn.datasets import load_iris  
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier  
from sklearn.tree import plot_tree 
from sklearn import metrics  
import matplotlib.pyplot as plt

iris = load_iris()
x = pd.DataFrame(iris.data, columns = iris.feature_names)
y = iris.target

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

clf = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', max_depth = 3, random_state = 0)
clf = clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)

print('Random Forest 1st Tree')

plt.figure(figsize=(15,10))
plot_tree(clf.estimators_[0], feature_names = iris.feature_names, class_names = list(iris.target_names), filled = True)
plt.show()

accuracy = metrics.accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

### Variable Explanation

- ***RandomForestClassifier -*** Built-in function that knows how to form the dataset into Random Forest
- ***load_iris -*** Built-in dataset which contains **n** number of rows of iris flowers specifications
- ***train_test_split -*** Split the dataset into training and testing dataset
- ***x -*** Array of input features/specification
- ***y -*** Array of output features/target
- ***random_state -*** Reproductability value 
- ***estimators -*** Represent number of decision trees it includes
- ***criterion -*** The criteria that the Algorithm forms the tree
- ***max_depth -*** Number of levels in the tree
